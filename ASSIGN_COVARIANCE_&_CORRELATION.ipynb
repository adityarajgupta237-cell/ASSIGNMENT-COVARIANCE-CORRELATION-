QUES.02. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

ANS- covariance shows how two variables move together. The sign of covariance indicates the direction of their relationship.

1. Positive Covariance (> 0)

A positive covariance means the two variables move in the same direction.

When one variable increases, the other also tends to increase.

When one decreases, the other tends to decrease.

Example:

- Income and expenditure

- Height and weight

This indicates a direct relationship between the variables.

2. Negative Covariance (< 0)

A negative covariance means the variables move in opposite directions.

When one variable increases, the other tends to decrease.

When one decreases, the other tends to increase.

Example:

- Price of a product and demand

- Speed of a vehicle and time taken to travel a fixed distance

This indicates an inverse relationship.

3. Zero Covariance (≈ 0)

A zero covariance means there is no linear relationship between the variables.

Changes in one variable do not systematically affect the other.

The variables may be independent or have no linear association.

- Example: Shoe size and intelligence (generally unrelated).

QUES.03.Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?

ANS.- Limitations of Covariance

Covariance indicates the direction of the relationship between two variables, but it has several limitations as a measure of association.

1.Depends on the Units of Measurement

Covariance is affected by the scale or units of the variables.

- If the units of measurement change (e.g., meters to centimeters), the covariance value also changes.

- Because of this, it is difficult to compare covariance values across different datasets.

2. Difficult to Interpret Magnitude

The value of covariance can range from negative infinity to positive infinity.

- There is no fixed range, so it is hard to determine the strength of the relationship between variables.

3. Not Standardized
Covariance is not normalized, meaning the magnitude depends on the variability of the variables themselves.

- Two datasets may have the same relationship but very different covariance values.

4. Cannot Easily Compare Relationships
Since covariance values depend on units and scale, it is not suitable for comparing relationships between different pairs of variables or across studies.

5. Limited Information About Strength

Covariance only indicates whether the relationship is positive or negative, but it does not clearly show how strong the relationship is.

Why Correlation is Preferred

Correlation overcomes many of these limitations because:

1.Unit-Free Measure
Correlation is a standardized form of covariance, so it does not depend on the units of measurement.

2.Fixed Range
The correlation coefficient lies between −1 and +1, making interpretation easier.

3.Shows Strength and Direction
It clearly indicates both the direction and the strength of the relationship between variables.

4.Allows Comparison
Because it is standardized, correlation values can be compared across different datasets and variables.

QUES.4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?

ANS.- 1. Pearson’s Correlation Coefficient (r)

Definition:
A statistical measure that evaluates the linear relationship between two continuous variables.

Key characteristics:

- Measures linear correlation.

- Uses the actual numerical values of the data.

- Assumes:

- Variables are normally distributed.

- Relationship is linear.

- Data are measured on interval or ratio scales.

- Sensitive to outliers.

Range:

−1≤r≤+1

+1 → Perfect positive linear relationship

−1 → Perfect negative linear relationship

0 → No linear relationship

Example:
Relationship between height and weight.

2. Spearman’s Rank Correlation Coefficient (ρ or rₛ)

Definition:
A non-parametric measure that evaluates the monotonic relationship between two variables using ranked data.

Key characteristics:

Uses ranks instead of raw data.

Measures monotonic relationships (not necessarily linear).

Does not require normal distribution.

Suitable for ordinal data.

Less affected by outliers.

Can we infer causation?

No, causation cannot be inferred from correlation alone.

A high correlation only shows that two variables move together, but it does not prove that one variable causes the other.

Reasons why correlation does not imply causation

1.Third variable (lurking variable):
Another variable may influence both X and Y.
Example: Ice cream sales and drowning incidents are correlated because hot weather affects both.

2.Reverse causality:
Y might influence X instead of X influencing Y.

3.Coincidence or spurious correlation:
Two variables may appear related purely by chance.
